```python
df1 = pd.DataFrame(raw_df)
```

```python
df1['Mileage'] = df1['Mileage'].str.extract('(\d+\.\d+)').astype(float)


mean_value = df1['Mileage'].mean()


df1['Mileage'].fillna(mean_value, inplace=True)
```

```python
df1['Engine'] = df1['Engine'].str.extract('(\d+)').astype(float)


mean_value = df1['Engine'].mean()


df1['Engine'].fillna(mean_value, inplace=True)
```

```python
df1['Power'] = df1['Power'].str.extract('(\d+\.\d+)').astype(float)


mean_value = df1['Power'].mean()


df1['Power'].fillna(mean_value, inplace=True)
```

```python
df1['Seats'].fillna(df1['Seats'].mode()[0], inplace=True)
```

```python
df1 = df1.drop(columns=["New_Price"])
```
